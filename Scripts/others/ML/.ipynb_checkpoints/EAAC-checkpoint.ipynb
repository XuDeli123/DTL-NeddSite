```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score,cross_validate
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re, os, sys
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from tqdm import tqdm
from sklearn.metrics import roc_curve,confusion_matrix,classification_report,auc
from sklearn import metrics
import math
from collections import Counter
from tqdm import tqdm
from scipy.stats import ttest_ind,ttest_rel
from sklearn.svm import SVR
```

```python
def EAAC(filepath, CodeType, windows=5):
    AA = 'ACDEFGHIKLMNPQRSTVWY'
    encodings = []
    dataframe = pd.read_csv(filepath)
    label = list(dataframe['label'])
    sequences = list(dataframe['sequence'])
    for i in sequences:
        sequence = re.sub('[^ACDEFGHIKLMNPQRSTVWYX]', 'X', ''.join(i).upper())
        code = []
        for j in range(len(sequence)):
            singlecode = []
            if j < len(sequence) and j + windows <= len(sequence):
                count = Counter(re.sub('X', '', sequence[j:j + windows]))
                for key in count:
                    count[key] = count[key] / windows
                for aa in AA:
                    singlecode.append(count[aa])
                code.append(singlecode)
        encodings.append(code)
    if CodeType == 1:
        return np.array(encodings).astype(np.float64), np.array(label).astype(np.float64)
    else:
        return np.array(encodings).astype(np.float64).reshape(len(encodings),-1), np.array(label).astype(np.float64)
```

```python
def plt_auc(number_fold,list_data)   :
    tprs = []
    aucs = []
    fprArray = []
    tprArray = []
    thresholdsArray = []
    mean_fpr = np.linspace(0, 1, 100)
    plt.figure(figsize=(8,8))
    plt.title('Validation ROC')
    colors = ["aqua","black","brown","orangered","peru","green","orange","darkviolet","teal","coral"]
    for i in range(number_fold):
        fpr, tpr, threshold = metrics.roc_curve(list_data[i][:,0], list_data[i][:,2])
        fprArray.append(fpr)
        tprArray.append(tpr)
        thresholdsArray.append(threshold)
        tprs.append(np.interp(mean_fpr, fpr, tpr))
        tprs[-1][0] = 0.0
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)
        roc_auc = metrics.auc(fpr, tpr)
        plt.plot(fpr, tpr, 'b', label = f'ROC Flod {i+1} :Val AUC = %0.3f' % roc_auc,color=colors[i])
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    plt.plot(mean_fpr, mean_tpr, color='blue',
                 label=r'Mean ROC (AUC = %0.4f $\pm$ %0.3f)' % (mean_auc, std_auc),
                 lw=2, alpha=.9)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    print(np.array(aucs).mean(),np.std(aucs))
```

```python
X,y=EAAC(r"C:\Users\CXL\Desktop\迁移学习整理\数据集\0.73\train.csv",0)
x_test,y_test=EAAC(r"C:\Users\CXL\Desktop\迁移学习整理\数据集\0.73\test.csv",0)
categories = sorted(set(y))
```

```python
# 调参，绘制学习曲线来调参n_estimators（对随机森林影响最大）
score_lt = []
# 每隔10步建立一个随机森林，获得不同n_estimators的得分
for i in tqdm(range(200,300,10)):
    rfc = RandomForestClassifier(n_estimators=i, bootstrap=True, oob_score=True, n_jobs=15)
    score = cross_val_score(rfc, X, y, cv=10).mean()
    score_lt.append(score)
score_max = max(score_lt)
```

```python
###粗调max_depth参数
ScoreAll = []
for i in tqdm(range(10,40,2)):
    rfc = RandomForestClassifier(n_estimators=290, bootstrap=True, oob_score=True, max_depth=i, n_jobs=15)
    score = cross_val_score(rfc, X, y, cv=10).mean()
    ScoreAll.append([i,score])
ScoreAll = np.array(ScoreAll)

max_score = np.where(ScoreAll==np.max(ScoreAll[:,1]))[0][0] ##找出最高得分对应的索引
print("最优参数以及最高得分:",ScoreAll[max_score])  
plt.figure(figsize=[20,5])
plt.plot(ScoreAll[:,0],ScoreAll[:,1])
plt.show()
```

```python
###调min_samples_split参数
ScoreAll = []
for i in tqdm(range(2,10)):
    rfc = RandomForestClassifier(n_estimators=290, bootstrap=True, oob_score=True,max_depth=32, min_samples_split=i, n_jobs=12)
    score = cross_val_score(rfc, X, y, cv=10).mean()
    ScoreAll.append([i,score])
ScoreAll = np.array(ScoreAll)

max_score = np.where(ScoreAll==np.max(ScoreAll[:,1]))[0][0] ##找出最高得分对应的索引
print("最优参数以及最高得分:",ScoreAll[max_score])  
plt.figure(figsize=[20,5])
plt.plot(ScoreAll[:,0],ScoreAll[:,1])
plt.show()
```

```python
###调min_samples_leaf参数
ScoreAll = []
for i in tqdm(range(1,15,1)):
    rfc = RandomForestClassifier(n_estimators=290, bootstrap=True, oob_score=True,max_depth=32,min_samples_split=7,min_samples_leaf=i, n_jobs=12)
    score = cross_val_score(rfc, X, y, cv=10).mean()
    ScoreAll.append([i,score])
ScoreAll = np.array(ScoreAll)

max_score = np.where(ScoreAll==np.max(ScoreAll[:,1]))[0][0] ##找出最高得分对应的索引
print("最优参数以及最高得分:",ScoreAll[max_score])  
plt.figure(figsize=[20,5])
plt.plot(ScoreAll[:,0],ScoreAll[:,1])
plt.show()
```

```python
categories = sorted(set(y))
prediction_result_cv = []
prediction_result_ind = []
folds = StratifiedKFold(10).split(X, y)
Acc=[]
Sn=[]
Sp=[]
Mcc=[]
Acc1=[]
Sn1=[]
Sp1=[]
Mcc1=[]
for i, (train, valid) in enumerate(folds):
    train_X, train_y = X[train], y[train]
    valid_X, valid_y = X[valid], y[valid]
    rfc_model = RandomForestClassifier(max_depth=32,min_samples_split=7,n_estimators=290,n_jobs=12)
    rfc = rfc_model.fit(train_X, train_y)
    
    tmp_result = np.zeros((len(valid_y), len(categories) + 1))
    tmp_result[:, 0], tmp_result[:, 1:] = valid_y, rfc.predict_proba(valid_X)
    prediction_result_cv.append(tmp_result)
    
    tmp_result1 = np.zeros((len(y_test), len(categories) + 1))
    tmp_result1[:, 0], tmp_result1[:, 1:] = y_test, rfc.predict_proba(x_test)
    prediction_result_ind.append(tmp_result1)
    
    y_pre=rfc.predict(valid_X)
    tn, fp, fn, tp = confusion_matrix(valid_y, y_pre).ravel()
    Acc.append((tp+tn)/(tn+fp+fn+tp))
    Sn.append(tp/(tp+fn))
    Sp.append(tn/(tn+fp))
    Mcc.append(((tp*tn)-(fn*fp))/(math.sqrt((tp+fn)*(tn+fp)*(tp+fp)*(tn+fn))))
    
    y_pre_ind=rfc.predict(x_test)
    tn1, fp1, fn1, tp1 = confusion_matrix(y_test, y_pre_ind).ravel()
    Acc1.append((tp1+tn1)/(tn1+fp1+fn1+tp1))
    Sn1.append(tp1/(tp1+fn1))
    Sp1.append(tn1/(tn1+fp1))
    Mcc1.append(((tp1*tn1)-(fn1*fp1))/(math.sqrt((tp1+fn1)*(tn1+fp1)*(tp1+fp1)*(tn1+fn1))))
```

```python
print(
    'Acc平方=%0.3f' % np.array(Acc).mean(),
    '±%0.3f' % np.std(Acc),
    '\n',
    'Sn平方=%0.3f' % np.array(Sn).mean(),
    '±%0.3f' % np.std(Sn),
    '\n',
    'Sp平方=%0.3f' % np.array(Sp).mean(),
    '±%0.3f' % np.std(Sp),
    '\n',
    'Mcc平方=%0.3f' % np.array(Mcc).mean(),
    '±%0.3f' %np.std(Mcc),'\n',
)
plt_auc(10,prediction_result_cv)
```

```python
print(
    'Acc平方=%0.3f' % np.array(Acc1).mean(),
    '±%0.3f' % np.std(Acc1),
    '\n',
    'Sn平方=%0.3f' % np.array(Sn1).mean(),
    '±%0.3f' % np.std(Sn1),
    '\n',
    'Sp平方=%0.3f' % np.array(Sp1).mean(),
    '±%0.3f' % np.std(Sp1),
    '\n',
    'Mcc平方=%0.3f' % np.array(Mcc1).mean(),
    '±%0.3f' %np.std(Mcc1),'\n',
)
plt_auc(10,prediction_result_ind)
```
